# Simulation of Two Stage Contagion in Social Network
See Katriel pdf here https://arxiv.org/pdf/1812.08867.pdf

## Model Parameters
 - $S_0$ Naive individuals who have not been exposed to the novel cultural motif.
 - $S_1$ Informed individuals who have been exposed to the novel cultural motif but do not espouse.  
 - $A$ Adopters of the novel cultural motif.
 - $R$ Adopters who abandon the novel cultural motif.
 
## Mechanisms involved in the model
 - Adopters randomly encounter other individuals at a rate $\beta$ contacts per unit of time.
 - Upon contact with an adopter, a naive individual becomes an adopter with probability $p \in (0,1)$ and otherwise becomes informed.
 - An informaed individual adopts upon encounter with an adopter.
 - Adopters abandon the social movement at a constant per capita rate $\gamma$. Adopters who abandon the social movement will not re-adopt.
 - Demographic turnover is not considered here as it is in Katriel's original model.

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
from random import *
%matplotlib inline

import time as timer
```

## Load the SNAP Network Data
Note I'm using Facebook data for now. The Twitter data is ~20 times larger and takes a long time to process/run simulations.

```python
# facebook network graph from https://snap.stanford.edu/data/egonets-Facebook.html
G_fb = nx.read_edgelist("./snap_network_data/facebook_combined.txt", create_using = nx.Graph(), nodetype = int)

## twitter network graph from https://snap.stanford.edu/data/ego-Twitter.html
# G_twtr = nx.read_edgelist("./snap_network_data/twitter_combined.txt", create_using = nx.Graph(), nodetype = int)
```

```python
# create network layout and visualize it
spring_pos = nx.spring_layout(G_fb)
plt.axis("off")
plt.figure(figsize=(12,8)) 
nx.draw_networkx(G_fb, pos = spring_pos, with_labels = False, node_size = 15)
plt.title('Facebook Network')
```

## Define Simulation Steps

```python
def initialize():
    global network, positions, nextNetwork
    
    network = G_fb

    positions = spring_pos

    # initialize a percent of the population as existing adopters
    # initialize the rest as naive
    for i in network.nodes:
        if random() < initial_pct_A:
            network.nodes[i]['state'] = adopter
        else:
            network.nodes[i]['state'] = naive

    nextNetwork = network.copy()


def observe():
    global n_adopter
    
    # count current total number of adopter nodes
    n_adopter = sum([1 for i in network.nodes if network.nodes[i]['state']==adopter])
    
    # add the current count to an accumulator time series 
    # so we can visualize it at the end of the sim
    n_adopter_ts.append(n_adopter)
    

def update():
    global network, nextNetwork

    # iterate over each node of the current network state
    for i in network.nodes:
        
        # find all adopters in the network, loop through their neighbors
        # and determine if the neighbor will adopt or become informed.
        if network.nodes[i]['state'] == adopter:
            
            # copy current network adopters to next network 
            nextNetwork.nodes[i]['state'] = adopter
            
            # loop through the adopter's neighbors
            for j in network.neighbors(i):
                
                # if the adopter encounters a naive, they convert them to adopter with probability p
                # otherwise, the neighbor becomes informed
                if network.nodes[j]['state'] == naive:
                    if random() < p:
                        nextNetwork.nodes[j]['state'] = adopter
                    else:
                        nextNetwork.nodes[j]['state'] = informed
                
                # if the adopter encounters an informed, they convert them to an adopter with probability p2
                if network.nodes[j]['state'] == informed:
                    if random() < p2:
                        nextNetwork.nodes[j]['state'] = adopter
                    else:
                        nextNetwork.nodes[j]['state'] = informed

            # adopters adandon a social movement with probability g
            if random() < g:
                nextNetwork.nodes[i]['state'] = abandoner


    # delete old network state and set the next network state as current
    del network
    network = nextNetwork.copy()
```

## Run the Sumulation

```python
EPOCHS = 50

# initial conditions
initial_pct_A = 0.001 # Note: 4039*0.001 = ~4.039 initial adopters at sim step 0
p = 0.01 # probability of instant adoption (S_0 -> A)
p2 = 1 # probability of subsequent adoption (S_1 -> A)
g = 0.5 # probablity of abandon (A -> R)

# node states
naive = 'S_0'
informed = 'S_1'
adopter = 'A'
abandoner = 'R'

# accumulator for n_adopters per step of the sim
n_adopter_ts = []

start = timer.time()
initialize()
for i in range(EPOCHS):
    observe()    
    update()

plt.figure(figsize=(12,6))
plt.plot(n_adopter_ts)
plt.title('Number of Adopters in Network')
plt.show()
end = timer.time()
print(end - start)
```
